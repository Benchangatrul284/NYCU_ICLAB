```python
import os
import hamming_codec
import numpy as np
```

```python
NUM_PATTERN = 1000
IP_BIT = 11
root_dir = 'ip_pattern'
np.random.seed(0)
```

```python
if root_dir not in os.listdir():
        os.mkdir(root_dir)

input_file_path = os.path.join(root_dir, 'input.txt')
output_file_path = os.path.join(root_dir, 'output.txt')
```

```python
def encode(data, num_bits):
    inserted_data0 = 0
    inserted_data1 = 0
    inserted_data2 = 0
    inserted_data3 = 0
    # compute the inserted value
    assert num_bits <= 11
    for i in range(num_bits):
        if (i == 0): # corresponding to the third bit (0011)
            if (data[i] == 1):
                inserted_data0 = not inserted_data0
                inserted_data1 = not inserted_data1
        if (i == 1): # corresponding to the fifth bit (0101)
            if (data[i] == 1):
                inserted_data0 = not inserted_data0
                inserted_data2 = not inserted_data2
        if (i == 2): # corresponding to the sixth bit (0110)
            if (data[i] == 1):
                inserted_data1 = not inserted_data1
                inserted_data2 = not inserted_data2
        if (i == 3): # corresponding to the seventh bit (0111)
            if (data[i] == 1):
                inserted_data0 = not inserted_data0
                inserted_data1 = not inserted_data1
                inserted_data2 = not inserted_data2
        if (i == 4): # corresponding to the ninth bit (1001)
            if (data[i] == 1):
                inserted_data0 = not inserted_data0
                inserted_data3 = not inserted_data3
        if (i == 5): # corresponding to the tenth bit (1010)
            if (data[i] == 1):
                inserted_data1 = not inserted_data1
                inserted_data3 = not inserted_data3
        if (i == 6): # corresponding to the eleventh bit (1011)
            if (data[i] == 1):
                inserted_data0 = not inserted_data0
                inserted_data1 = not inserted_data1
                inserted_data3 = not inserted_data3
        if (i == 7):
            if (data[i] == 1): # corresponding to the twelfth bit (1100)
                inserted_data2 = not inserted_data2
                inserted_data3 = not inserted_data3
        if (i == 8):
            if (data[i] == 1): # corresponding to the thirteenth bit (1101)
                inserted_data0 = not inserted_data0
                inserted_data2 = not inserted_data2
                inserted_data3 = not inserted_data3
        if (i == 9):
            if (data[i] == 1): # corresponding to the fourteenth bit (1110)
                inserted_data1 = not inserted_data1
                inserted_data2 = not inserted_data2
                inserted_data3 = not inserted_data3
        if (i == 10): 
            if (data[i] == 1): # corresponding to the fifteenth bit (1111)
                inserted_data0 = not inserted_data0
                inserted_data1 = not inserted_data1
                inserted_data2 = not inserted_data2
                inserted_data3 = not inserted_data3
                
        # print(f"i = {i}")
        # print(f"data[i] = {data[i]}")
        # print(f"inserted_data0 = {inserted_data0}")
        # print(f"inserted_data1 = {inserted_data1}")
        # print(f"inserted_data2 = {inserted_data2}")
        # print(f"inserted_data3 = {inserted_data3}")
    encoded_data = list(data)
    encoded_data.insert(0,inserted_data0)
    encoded_data.insert(1,inserted_data1)
    encoded_data.insert(3,inserted_data2)
    encoded_data.insert(7,inserted_data3)
    return np.array(encoded_data)
```

```python
# decode
def decode(data, num_bits):
    inserted_data0 = 0
    inserted_data1 = 0
    inserted_data2 = 0
    inserted_data3 = 0
    assert num_bits <= 15
    for i in range(num_bits):
        if (i == 0):
            if (data[i] == 1): # corresponding to the first bit (0001)
                inserted_data0 = not inserted_data0
        if (i == 1):
            if (data[i] == 1): # corresponding to the second bit (0010)
                inserted_data1 = not inserted_data1
        if (i == 2): # corresponding to the third bit (0011)
            if (data[i] == 1):
                inserted_data0 = not inserted_data0
                inserted_data1 = not inserted_data1
        if (i == 3): # corresponding to the fourth bit (0100)
            if (data[i] == 1):
                inserted_data2 = not inserted_data2
        if (i == 4): # corresponding to the fifth bit (0101)
            if (data[i] == 1):
                inserted_data0 = not inserted_data0
                inserted_data2 = not inserted_data2
        if (i == 5): # corresponding to the sixth bit (0110)
            if (data[i] == 1):
                inserted_data1 = not inserted_data1
                inserted_data2 = not inserted_data2
        if (i == 6): # corresponding to the seventh bit (0111)
            if (data[i] == 1):
                inserted_data0 = not inserted_data0
                inserted_data1 = not inserted_data1
                inserted_data2 = not inserted_data2
        if (i == 7): # corresponding to the eighth bit (1000)
            if (data[i] == 1):
                inserted_data3 = not inserted_data3
        if (i == 8): # corresponding to the ninth bit (1001)
            if (data[i] == 1):
                inserted_data0 = not inserted_data0
                inserted_data3 = not inserted_data3
        if (i == 9): # corresponding to the tenth bit (1010)
            if (data[i] == 1):
                inserted_data1 = not inserted_data1
                inserted_data3 = not inserted_data3
        if (i == 10): # corresponding to the eleventh bit (1011)
            if (data[i] == 1):
                inserted_data0 = not inserted_data0
                inserted_data1 = not inserted_data1
                inserted_data3 = not inserted_data3
        if (i == 11):
            if (data[i] == 1): # corresponding to the twelfth bit (1100)
                inserted_data2 = not inserted_data2
                inserted_data3 = not inserted_data3
        if (i == 12):
            if (data[i] == 1): # corresponding to the thirteenth bit (1101)
                inserted_data0 = not inserted_data0
                inserted_data2 = not inserted_data2
                inserted_data3 = not inserted_data3
        if (i == 13):
            if (data[i] == 1): # corresponding to the fourteenth bit (1110)
                inserted_data1 = not inserted_data1
                inserted_data2 = not inserted_data2
                inserted_data3 = not inserted_data3
        if (i == 14):
            if (data[i] == 1): # corresponding to the fifteenth bit (1111)
                inserted_data0 = not inserted_data0
                inserted_data1 = not inserted_data1
                inserted_data2 = not inserted_data2
                inserted_data3 = not inserted_data3
                
        # print(f'i = {i}')
        # print("inserted_data0: ",inserted_data0)
        # print("inserted_data1: ",inserted_data1)
        # print("inserted_data2: ",inserted_data2)
        # print("inserted_data3: ",inserted_data3)
        
        
    position = inserted_data0 + inserted_data1*2 + inserted_data2*4 + inserted_data3*8
    if (position != 0):
        print(f"Error detected at position {position}")
        data[position-1] = not data[position-1]
    else:
        # print("No error detected")
        pass
        
    # remove the inserted value
    decoded_data = list(data)
    decoded_data.pop(7)
    decoded_data.pop(3)
    decoded_data.pop(1)
    decoded_data.pop(0)
    return np.array(decoded_data)
    
    
```

```python
input_file = open(input_file_path, 'w')
output_file = open(output_file_path, 'w')
error_rate = 1

for i in range(NUM_PATTERN):
    input_file.write(f"{i}\n")
    output_file.write(f"{i}\n")
    data = (np.random.randint(0, 2, IP_BIT))
    # print("data = ", data)
    encoded_data = encode(data, IP_BIT)
    print("Pattern ", i)
    print("encoded_data = ", encoded_data)
    # intentionally introduce error
    if np.random.rand() < error_rate:
        error_position = np.random.randint(0, IP_BIT)
        encoded_data[error_position] = not encoded_data[error_position]
        # print("encoded_data = ", encoded_data)
        # input_file.write(''.join(map(str, encoded_data)) + '\n')
    
    print("encoded_data = ", encoded_data)
    input_file.write(''.join(map(str, encoded_data)) + '\n')
    
    decoded_data = decode(encoded_data, IP_BIT+4)
    
    output_file.write(''.join(map(str, decoded_data)) + '\n')
    assert np.array_equal(data, decoded_data)
    
input_file.close()
output_file.close()
    
```
