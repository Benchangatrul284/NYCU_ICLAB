```python
import os
import numpy as np
```

```python
NUM_PATTERN = 100000
root_dir = 'lab6_pattern'
np.random.seed(0)
```

```python
if root_dir not in os.listdir():
        os.mkdir(root_dir)

input_file_path = os.path.join(root_dir, 'input.txt')
mode_file_path = os.path.join(root_dir, 'mode.txt')
output_file_path = os.path.join(root_dir, 'output.txt')
```

```python
# decode
def decode(data, num_bits):
    inserted_data0 = 0
    inserted_data1 = 0
    inserted_data2 = 0
    inserted_data3 = 0
    assert num_bits <= 15
    for i in range(num_bits):
        if (i == 0):
            if (data[i] == 1): # corresponding to the first bit (0001)
                inserted_data0 = not inserted_data0
        if (i == 1):
            if (data[i] == 1): # corresponding to the second bit (0010)
                inserted_data1 = not inserted_data1
        if (i == 2): # corresponding to the third bit (0011)
            if (data[i] == 1):
                inserted_data0 = not inserted_data0
                inserted_data1 = not inserted_data1
        if (i == 3): # corresponding to the fourth bit (0100)
            if (data[i] == 1):
                inserted_data2 = not inserted_data2
        if (i == 4): # corresponding to the fifth bit (0101)
            if (data[i] == 1):
                inserted_data0 = not inserted_data0
                inserted_data2 = not inserted_data2
        if (i == 5): # corresponding to the sixth bit (0110)
            if (data[i] == 1):
                inserted_data1 = not inserted_data1
                inserted_data2 = not inserted_data2
        if (i == 6): # corresponding to the seventh bit (0111)
            if (data[i] == 1):
                inserted_data0 = not inserted_data0
                inserted_data1 = not inserted_data1
                inserted_data2 = not inserted_data2
        if (i == 7): # corresponding to the eighth bit (1000)
            if (data[i] == 1):
                inserted_data3 = not inserted_data3
        if (i == 8): # corresponding to the ninth bit (1001)
            if (data[i] == 1):
                inserted_data0 = not inserted_data0
                inserted_data3 = not inserted_data3
        if (i == 9): # corresponding to the tenth bit (1010)
            if (data[i] == 1):
                inserted_data1 = not inserted_data1
                inserted_data3 = not inserted_data3
        if (i == 10): # corresponding to the eleventh bit (1011)
            if (data[i] == 1):
                inserted_data0 = not inserted_data0
                inserted_data1 = not inserted_data1
                inserted_data3 = not inserted_data3
        if (i == 11):
            if (data[i] == 1): # corresponding to the twelfth bit (1100)
                inserted_data2 = not inserted_data2
                inserted_data3 = not inserted_data3
        if (i == 12):
            if (data[i] == 1): # corresponding to the thirteenth bit (1101)
                inserted_data0 = not inserted_data0
                inserted_data2 = not inserted_data2
                inserted_data3 = not inserted_data3
        if (i == 13):
            if (data[i] == 1): # corresponding to the fourteenth bit (1110)
                inserted_data1 = not inserted_data1
                inserted_data2 = not inserted_data2
                inserted_data3 = not inserted_data3
        if (i == 14):
            if (data[i] == 1): # corresponding to the fifteenth bit (1111)
                inserted_data0 = not inserted_data0
                inserted_data1 = not inserted_data1
                inserted_data2 = not inserted_data2
                inserted_data3 = not inserted_data3
                
        # print(f'i = {i}')
        # print("inserted_data0: ",inserted_data0)
        # print("inserted_data1: ",inserted_data1)
        # print("inserted_data2: ",inserted_data2)
        # print("inserted_data3: ",inserted_data3)
        
        
    position = inserted_data0 + inserted_data1*2 + inserted_data2*4 + inserted_data3*8
    if (position != 0):
        # print(f"Error detected at position {position}")
        data[position-1] = not data[position-1]
    else:
        # print("No error detected")
        pass
        
    # remove the inserted value
    decoded_data = list(data)
    decoded_data.pop(7)
    decoded_data.pop(3)
    decoded_data.pop(1)
    decoded_data.pop(0)
    return np.array(decoded_data)
```

```python
def encode(data, num_bits):
    inserted_data0 = 0
    inserted_data1 = 0
    inserted_data2 = 0
    inserted_data3 = 0
    # compute the inserted value
    assert num_bits <= 11
    for i in range(num_bits):
        if (i == 0): # corresponding to the third bit (0011)
            if (data[i] == 1):
                inserted_data0 = not inserted_data0
                inserted_data1 = not inserted_data1
        if (i == 1): # corresponding to the fifth bit (0101)
            if (data[i] == 1):
                inserted_data0 = not inserted_data0
                inserted_data2 = not inserted_data2
        if (i == 2): # corresponding to the sixth bit (0110)
            if (data[i] == 1):
                inserted_data1 = not inserted_data1
                inserted_data2 = not inserted_data2
        if (i == 3): # corresponding to the seventh bit (0111)
            if (data[i] == 1):
                inserted_data0 = not inserted_data0
                inserted_data1 = not inserted_data1
                inserted_data2 = not inserted_data2
        if (i == 4): # corresponding to the ninth bit (1001)
            if (data[i] == 1):
                inserted_data0 = not inserted_data0
                inserted_data3 = not inserted_data3
        if (i == 5): # corresponding to the tenth bit (1010)
            if (data[i] == 1):
                inserted_data1 = not inserted_data1
                inserted_data3 = not inserted_data3
        if (i == 6): # corresponding to the eleventh bit (1011)
            if (data[i] == 1):
                inserted_data0 = not inserted_data0
                inserted_data1 = not inserted_data1
                inserted_data3 = not inserted_data3
        if (i == 7):
            if (data[i] == 1): # corresponding to the twelfth bit (1100)
                inserted_data2 = not inserted_data2
                inserted_data3 = not inserted_data3
        if (i == 8):
            if (data[i] == 1): # corresponding to the thirteenth bit (1101)
                inserted_data0 = not inserted_data0
                inserted_data2 = not inserted_data2
                inserted_data3 = not inserted_data3
        if (i == 9):
            if (data[i] == 1): # corresponding to the fourteenth bit (1110)
                inserted_data1 = not inserted_data1
                inserted_data2 = not inserted_data2
                inserted_data3 = not inserted_data3
        if (i == 10): 
            if (data[i] == 1): # corresponding to the fifteenth bit (1111)
                inserted_data0 = not inserted_data0
                inserted_data1 = not inserted_data1
                inserted_data2 = not inserted_data2
                inserted_data3 = not inserted_data3
                
        # print(f"i = {i}")
        # print(f"data[i] = {data[i]}")
        # print(f"inserted_data0 = {inserted_data0}")
        # print(f"inserted_data1 = {inserted_data1}")
        # print(f"inserted_data2 = {inserted_data2}")
        # print(f"inserted_data3 = {inserted_data3}")
    encoded_data = list(data)
    encoded_data.insert(0,inserted_data0)
    encoded_data.insert(1,inserted_data1)
    encoded_data.insert(3,inserted_data2)
    encoded_data.insert(7,inserted_data3)
    return np.array(encoded_data)
```

```python
input_file = open(input_file_path, 'w')
mode_file = open(mode_file_path, 'w')
output_file = open(output_file_path, 'w')
error_rate = 1
for i in range(NUM_PATTERN):
    input_file.write(f"{i}\n")
    mode_file.write(f"{i}\n")
    output_file.write(f"{i}\n")
    print("Pattern ", i)
    for j in range(16):
        pixel = (np.random.randint(0, 2, 11))
        print("pixel = ", pixel)
        encoded_pixel = encode(pixel, 11)
        # print("encoded_pixel = ", encoded_pixel)
        # intentionally introduce error
        if np.random.rand() < error_rate:
            error_position = np.random.randint(0, len(encoded_pixel))
            encoded_pixel[error_position] = not encoded_pixel[error_position]
    
        input_file.write(''.join(map(str, encoded_pixel)) + '\n')
    
    
    # mode = (np.random.randint(0, 2, 5))
    mode_possible_array = []
    mode_possible_array.append(np.array([0,0,1,0,0]))
    mode_possible_array.append(np.array([0,0,1,1,0]))
    mode_possible_array.append(np.array([1,0,1,1,0]))
    mode = mode_possible_array[np.random.randint(0, 3)]
    # mode = mode_possible_array[2]
    print("mode = ", mode)
    encoded_mode = encode(mode, 5)
    if np.random.rand() < error_rate:
        error_position = np.random.randint(0, len(encoded_mode))
        encoded_mode[error_position] = not encoded_mode[error_position]
    mode_file.write(''.join(map(str, encoded_mode)) + '\n')
    
input_file.close()
mode_file.close()
```

```python
def todecimal(x, bits):
    assert len(x) <= bits
    n = int(x, 2)
    s = 1 << (bits - 1)
    return (n & s - 1) - (n & s)

output_file = open(output_file_path, 'w')
mode_file = open(mode_file_path, 'r')
input_file = open(input_file_path, 'r')

# cacluate the result
for pat in range(NUM_PATTERN):
    output_file.write(f'{pat}\n')
    # read the files
    print(f'Pattern {mode_file.readline().strip()} for mode file')
    print(f'Pattern {input_file.readline().strip()} for input file')
    # read the mode
    mode = list(map(int, mode_file.readline().strip()))
    # decode the mode
    decoded_mode = decode(mode, 5+4)
    print(f'decoded_mode = {decoded_mode}')

    # read the data
    pixel = np.zeros((16, 11), dtype=int)
    for i in range(16):
        encoded_pixel = np.array(list(map(int, input_file.readline().strip())))
        decoded_pixel = decode(encoded_pixel, 11+4)
        pixel[i] = np.array(decoded_pixel)
   
    print(f'pixel = \n{pixel}')
    # calculate the result
    # pixel is a signed integer of 11 bits
    pixel_dec = np.zeros((16), dtype=int)
    for i in range(16):
        pixel_dec[i] = todecimal(''.join(map(str, pixel[i])), 11)
    
    print("Decoded pixel")
    for i in range(4): 
        print(f'{pixel_dec[i*4]} {pixel_dec[i*4+1]} {pixel_dec[i*4+2]} {pixel_dec[i*4+3]}')
    
    if (decoded_mode[0] == 0 and decoded_mode[1] == 0 and decoded_mode[2] == 1 and decoded_mode[3] == 0 and decoded_mode[4] == 0):
        det_size = 2
    elif (decoded_mode[0] == 0 and decoded_mode[1] == 0 and decoded_mode[2] == 1 and decoded_mode[3] == 1 and decoded_mode[4] == 0):
        det_size = 3
    elif (decoded_mode[0] == 1 and decoded_mode[1] == 0 and decoded_mode[2] == 1 and decoded_mode[3] == 1 and decoded_mode[4] == 0):
        det_size = 4
    else:
        print("Invalid mode")
        break
    
    if (det_size == 2):
        # pixel is a 4 by 4 matrix
        out_matrix = np.zeros(9)
        out_matrix[0] = np.linalg.det(np.array([[pixel_dec[0], pixel_dec[1]], [pixel_dec[4], pixel_dec[5]]]))
        out_matrix[1] = np.linalg.det(np.array([[pixel_dec[1], pixel_dec[2]], [pixel_dec[5], pixel_dec[6]]]))
        out_matrix[2] = np.linalg.det(np.array([[pixel_dec[2], pixel_dec[3]], [pixel_dec[6], pixel_dec[7]]]))
        out_matrix[3] = np.linalg.det(np.array([[pixel_dec[4], pixel_dec[5]], [pixel_dec[8], pixel_dec[9]]]))
        out_matrix[4] = np.linalg.det(np.array([[pixel_dec[5], pixel_dec[6]], [pixel_dec[9], pixel_dec[10]]]))
        out_matrix[5] = np.linalg.det(np.array([[pixel_dec[6], pixel_dec[7]], [pixel_dec[10], pixel_dec[11]]]))
        out_matrix[6] = np.linalg.det(np.array([[pixel_dec[8], pixel_dec[9]], [pixel_dec[12], pixel_dec[13]]]))
        out_matrix[7] = np.linalg.det(np.array([[pixel_dec[9], pixel_dec[10]], [pixel_dec[13], pixel_dec[14]]]))
        out_matrix[8] = np.linalg.det(np.array([[pixel_dec[10], pixel_dec[11]], [pixel_dec[14], pixel_dec[15]]]))
        print(f'out_matrix = \n{out_matrix}')
        # convert the result into 23 bits signed binary number
        out_matrix_bin = np.zeros(9)
        for i in range(9):
            binary_str = np.binary_repr(int(round(out_matrix[i])), 23)
            print(f'binary_str = {binary_str}')
            out_matrix_bin[i] = int(binary_str, 2)
            output_file.write(''.join(map(str, binary_str)))
        output_file.write('\n')
    elif (det_size == 3):
        # pixel is a 4 by 4 matrix
        out_matrix = np.zeros(4)
        out_matrix[0] = np.linalg.det(np.array([[pixel_dec[0], pixel_dec[1], pixel_dec[2]], [pixel_dec[4], pixel_dec[5], pixel_dec[6]], [pixel_dec[8], pixel_dec[9], pixel_dec[10]]]))
        out_matrix[1] = np.linalg.det(np.array([[pixel_dec[1], pixel_dec[2], pixel_dec[3]], [pixel_dec[5], pixel_dec[6], pixel_dec[7]], [pixel_dec[9], pixel_dec[10], pixel_dec[11]]]))
        out_matrix[2] = np.linalg.det(np.array([[pixel_dec[4], pixel_dec[5], pixel_dec[6]], [pixel_dec[8], pixel_dec[9], pixel_dec[10]], [pixel_dec[12], pixel_dec[13], pixel_dec[14]]]))
        out_matrix[3] = np.linalg.det(np.array([[pixel_dec[5], pixel_dec[6], pixel_dec[7]], [pixel_dec[9], pixel_dec[10], pixel_dec[11]], [pixel_dec[13], pixel_dec[14], pixel_dec[15]]]))
        print(f'out_matrix = \n{out_matrix}')
        # convert the result into 23 bits signed binary number
        out_matrix_bin = np.zeros(4)
        for i in range(4):
            binary_str = np.binary_repr(int(round(out_matrix[i])), 51)
            print(f'binary_str = {binary_str}')
            out_matrix_bin[i] = int(binary_str, 2)
            if (i == 0):
                output_file.write('000')
            output_file.write(''.join(map(str, binary_str)))
        output_file.write('\n')
        
    elif (det_size == 4):
        out_matrix = np.linalg.det(np.array([[pixel_dec[0], pixel_dec[1], pixel_dec[2], pixel_dec[3]], [pixel_dec[4], pixel_dec[5], pixel_dec[6], pixel_dec[7]], [pixel_dec[8], pixel_dec[9], pixel_dec[10], pixel_dec[11]], [pixel_dec[12], pixel_dec[13], pixel_dec[14], pixel_dec[15]]]))
        print(f'out_matrix = \n{out_matrix}')
        # print intermediate result
        print("====stage 1:===========")
        print("det value:")
        det_value = np.linalg.det(np.array([[pixel_dec[1], pixel_dec[2], pixel_dec[3]], [pixel_dec[5], pixel_dec[6], pixel_dec[7]], [pixel_dec[9], pixel_dec[10], pixel_dec[11]]]))
        print(f'det_value = {det_value}')
        coeff = pixel_dec[12]
        print(f'coeff = {coeff}')
        print("result:")
        print(det_value*coeff)
        print("====stage 2:===========")
        print("det value:")
        det_value = np.linalg.det(np.array([[pixel_dec[0], pixel_dec[2], pixel_dec[3]], [pixel_dec[4], pixel_dec[6], pixel_dec[7]], [pixel_dec[8], pixel_dec[10], pixel_dec[11]]]))
        print(f'det_value = {det_value}')
        coeff = pixel_dec[13]
        print(f'coeff = {coeff}')
        print("result:")
        print(det_value*coeff)
        print("====stage 3:===========")
        print("det value:")
        det_value = np.linalg.det(np.array([[pixel_dec[0], pixel_dec[1], pixel_dec[3]], [pixel_dec[4], pixel_dec[5], pixel_dec[7]], [pixel_dec[8], pixel_dec[9], pixel_dec[11]]]))
        print(f'det_value = {det_value}')
        coeff = pixel_dec[14]
        print(f'coeff = {coeff}')
        print("result:")
        print(det_value*coeff)
        print("====stage 4:===========")
        print("det value:")
        det_value = np.linalg.det(np.array([[pixel_dec[0], pixel_dec[1], pixel_dec[2]], [pixel_dec[4], pixel_dec[5], pixel_dec[6]], [pixel_dec[8], pixel_dec[9], pixel_dec[10]]]))
        print(f'det_value = {det_value}')
        coeff = pixel_dec[15]
        print(f'coeff = {coeff}')
        print("result:")
        print(det_value*coeff)
        # convert the result into 207 bits signed binary number
        out_matrix_bin = np.binary_repr(int(round(out_matrix)), 207)
        print(f'out_matrix_bin = {out_matrix_bin}')
        output_file.write(''.join(map(str, out_matrix_bin)))
        output_file.write('\n')
        
input_file.close()
mode_file.close()
output_file.close()
```
